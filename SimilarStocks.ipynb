```python
import re
import requests
import sys
import pandas as pd
import json
```

```python
with open('relatedCompanies.json') as tickers:
    dict_tickers = json.load(tickers)
    
relcos = list(set(dict_tickers['relcos']))

for item in relcos:
    if ' ' not in item:
        print(item)
```

```python
df = pd.read_csv('SP400.csv', encoding = "ISO-8859-1")

count = 0

relatedCompanies = {'relcos':[]}

for ticker in df['ticker']:
    try:
        url = 'https://www.google.com/finance?q=NYSE:' + ticker
        urlRelated = 'https://www.google.com/finance/related?q=NYSE:' + ticker 

        r = requests.get(url)
        r = r.text

        matches = re.findall('navmenu(.*?)Related companies', r, re.DOTALL)
        matches = re.findall(ticker + '(.*?)">', matches[0], re.DOTALL)

        r = requests.get(urlRelated + matches[0])
        r = r.text

        matches = re.findall('google.finance.data = (.*?);', r, re.DOTALL)

        for m in re.finditer('values:\["(.*?)",', matches[0], re.DOTALL):
            relatedCompanies['relcos'].append(m.group(1))

        count += 1
        print(ticker, count)
        
    except:
        print(ticker, count, 'Failed')
        count += 1
        pass
        
#json_str = json.dumps(relatedCompanies)
#print(json_str)
with open('relatedCompanies.json', 'w') as outfile:
    json.dump(relatedCompanies, outfile)
```
